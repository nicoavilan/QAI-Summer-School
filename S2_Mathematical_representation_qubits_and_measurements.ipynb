## **Mathematical representation of qubits and their measurements**


## **Dirac's bra-ket notation**:

A qubit is a two-level quantum system described by a state vector in a 2D complex Hilbert space.

*Bra-Ket*: $\langle \psi_1 | \psi_2\rangle$

In quantum computing, the state of a qubit is commonly written using **ket** notation as:

$$ |\psi\rangle = \alpha|0\rangle + \beta|1\rangle $$

$|0\rangle$ and $|1\rangle$ are the computational basis states, represented as:

$$ |0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, \quad |1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$$

where $ \alpha$, $\beta$ $\in \mathbb{C}$, and satisfy the normalization condition: $ |\alpha|^2 + |\beta|^2 = 1 $.

Take into account that $ |\alpha|^2 = \alpha^*\alpha$ for $\alpha\in \mathbb{C}$.


**Bra**: The conjugate transpose of $|\psi\rangle$ is denoted by $\langle\psi| = \alpha \langle 0| + \beta \langle 1|$, where $\langle 0| = (1\quad 0)$, $\langle1| = (0\quad 1)$.

For the state vector in matrix form  $|\psi\rangle = \begin{pmatrix} \alpha \\ \beta \end{pmatrix}$, the inner product $\langle\psi|\psi\rangle = |\alpha|^2 + |\beta|^2 = 1$ ensures normalization.

## **Bloch Sphere**

The qubit state can be visualized on the Bloch sphere, where a pure state is: $|\psi\rangle = \cos\left(\frac{\theta}{2}\right)|0\rangle + e^{i\phi}\sin\left(\frac{\theta}{2}\right)|1\rangle$

$\theta$ and $\phi$ are defined using the standard notation commonly adopted by physicists.

$\theta$: Polar angle (0 to $\pi$).

$\phi$: Azimuthal angle (0 to $2\pi$).

Bloch vector: $ (x, y, z) = (\sin\theta \cos\phi, \sin\theta \sin\phi, \cos\theta)$.


image.png

**Example**: A qubit in an equal superposition (50% chance of $|0\rangle $, 50% chance of $|1\rangle$) is: $$|\psi\rangle = \frac{1}{\sqrt{2}}|0\rangle + \frac{1}{\sqrt{2}}|1\rangle = \begin{pmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{pmatrix}$$

We evaluate the probabilities using **Born's rule**, which states that the probability of measuring a quantum system in a particular state is given by the squared magnitude of the corresponding amplitude in the state vector.

## **Quantum State Evolution**

Quantum states are transformed by quantum gates-reversible operations that modify the state of qubits. Unlike classical logic gates, which act on bits in definite states (0 or 1), quantum gates operate on qubits that can exist in superpositions of both basis states. These operations are defined within a complex vector space known as Hilbert space.

Mathematically, quantum gates are represented by unitary operators—matrices $U$ that satisfy the condition $U^\dagger U=I$, where $U^\dagger$ is the *conjugate transpose* of $U$ and $I$ is the identity matrix. When a gate is applied, the quantum state evolves according to the rule $|\psi'\rangle = U| \psi\rangle $.

**Example**: *Hadamard Gate*

The Hadamard gate creates superposition:

$$ H = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix} $$

For $|0\rangle$:
$$ H|0\rangle = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \begin{pmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{pmatrix} = \frac{1}{\sqrt{2}}|0\rangle + \frac{1}{\sqrt{2}}|1\rangle $$

**The Hadamard Basis**

The Hadamard operator allows us to define a new vector basis called the Hadamard basis or the X-basis, given by

$$ |+\rangle = H|0\rangle = \frac{1}{\sqrt{2}}|0\rangle + \frac{1}{\sqrt{2}}|1\rangle $$
$$ |-\rangle = H|1\rangle = \frac{1}{\sqrt{2}}|0\rangle - \frac{1}{\sqrt{2}}|1\rangle $$

## **Creating an Unequal Superposition with $ R_y $ Concept**

Hadamard gate transforms the basis states into an equal amplitud superposition, nevertheless it is important to see that this is not the only posibility. General rotations allows to create unequal aplitud supoerpositions.

Use the $ R_y $ rotation gate to create a superposition with unequal amplitudes, illustrating how rotation angles control probability distributions.

*Mathematical Background*:

The $ R_y(\theta) $ gate rotates the qubit state around the y-axis of the Bloch sphere:
$$R_y(\theta) = \begin{pmatrix} \cos(\theta/2) & -\sin(\theta/2) \\ \sin(\theta/2) & \cos(\theta/2) \end{pmatrix}$$


<font color="red"> **Does the operator $R_y(\theta)$ satisfy the conditions of unitarity?** </font>



For $ |0\rangle $:
$$R_y(\theta)|0\rangle = \cos(\theta/2)|0\rangle + \sin(\theta/2)|1\rangle$$
Choose $ \theta = \frac{\pi}{3} $ to get:
$$|\psi\rangle = \cos(\pi/6)|0\rangle + \sin(\pi/6)|1\rangle = \frac{\sqrt{3}}{2}|0\rangle + \frac{1}{2}|1\rangle$$

**Probabilities:**

$$ P(|0\rangle) = \left|\frac{\sqrt{3}}{2}\right|^2 = \frac{3}{4}, \qquad  P(|1\rangle) = \left|\frac{1}{2}\right|^2 = \frac{1}{4} $$

For $\theta = \frac{\pi}{3} $, the probability of measuring the state in $ |0\rangle $ is 75\% and in $ |1\rangle $ is 25\%.

Bloch vector: Lies on the xz-plane, tilted toward the north pole.

## **Phase Manipulation with $ S $ Gate**

Use the $ S $ gate to introduce a relative phase, demonstrating phase's role in quantum states without altering measurement probabilities in the computational basis.


**Mathematical Background:**

The $ S $ gate applies a $ \frac{\pi}{2} $ phase to $ |1\rangle $:
$$S = \begin{pmatrix} 1 & 0 \\ 0 & i \end{pmatrix}$$

Start with
$$ |+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) $$ created by a Hadamard-like state preparation $ |+\rangle = H|0\rangle $.


Apply $ S $:

$$S|+\rangle = S \cdot \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) = \frac{1}{\sqrt{2}}(|0\rangle + i|1\rangle) = |i\rangle$$

**The Pauli-Y Basis**

The Pauli-Y operator is given by:

$$ Y = \begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix} $$

The Y-basis is given by

$$ |i\rangle = S|+\rangle = \frac{1}{\sqrt{2}}|0\rangle + \frac{i}{\sqrt{2}}|1\rangle $$
$$ |-i\rangle = S|-\rangle = \frac{1}{\sqrt{2}}|0\rangle - \frac{i}{\sqrt{2}}|1\rangle $$

The $ S $ gate introduces a phase, transforming $ \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) $ to $ \frac{1}{\sqrt{2}}(|0\rangle + i|1\rangle) $.
$$ \frac{1}{\sqrt{2}}(|0\rangle + i|1\rangle) = \frac{1}{\sqrt{2}}(|0\rangle + e^{i\pi/2}|1\rangle) $$

This *rotates the Bloch vector* to the y-axis, but measurement probabilities remain 50%.

**Probabilities in computational basis**:

$ P(|0\rangle) = P(|1\rangle) = \frac{1}{2} $.

## **Phase Manipulation with $ T $ Gate**
Use the $ T $ gate to introduce a $ \frac{\pi}{4} $ phase, showing how phase affects the state without changing computational basis probabilities.

**Mathematical Background**:

The $ T $ gate applies a $ \frac{\pi}{4} $ phase to $ |1\rangle $:

$$T = \begin{pmatrix} 1 & 0 \\ 0 & e^{i\pi/4} \end{pmatrix}$$

Start with $ |+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) $
$$T|+\rangle = T \cdot \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) = \frac{1}{\sqrt{2}}(|0\rangle + e^{i\pi/4}|1\rangle)$$

Probabilities:
$$P(|0\rangle) = \left| \frac{1}{\sqrt{2}} \right|^2 = \frac{1}{2}, \quad P(|1\rangle) = \left| \frac{e^{i\pi/4}}{\sqrt{2}} \right|^2 = \frac{1}{2}$$


The $ T $ gate applies a $ \frac{\pi}{4} $ phase to $ |1\rangle $, transforming $ \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) $ to $ \frac{1}{\sqrt{2}}(|0\rangle + e^{i\pi/4}|1\rangle) $. This rotates the Bloch vector to the xy-plane at 45°, but measurement probabilities in the computational basis remain 50%.
